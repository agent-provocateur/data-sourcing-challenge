### Import Required Libraries and Set Up Environment Variables

```python
# Dependencies
import requests
import time
from dotenv import load_dotenv
import os
import pandas as pd
import json
from colorama import Fore, Style
```

```python
# Set environment variables from the .env in the local environment
load_dotenv()

nyt_api_key = os.getenv("NYT_API_KEY")
tmdb_api_key = os.getenv("TMDB_API_KEY")
```

### Access the New York Times API

```python
# Set the base URL
url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Filter for movie reviews with "love" in the headline
# section_name should be "Movies"
# type_of_material should be "Review"
filter_query = 'section_name:"Movies" AND type_of_material:"Review" AND headline:"love"'

# Use a sort filter, sort by newest
sort = "newest"

# Select the following fields to return:
# headline, web_url, snippet, source, keywords, pub_date, byline, word_count
field_list = "headline,web_url,snippet,source,keywords,pub_date,byline,word_count"

# Search for reviews published between a begin and end date
begin_date = "20130101"
end_date = "20230531"

# Build URL
movie_url = f'{url}fq={filter_query}&begin_date={begin_date}&end_date={end_date}&sort={sort}&fl={field_list}&api-key={nyt_api_key}'
```

```python
# Create an empty list to store the reviews
reviews_list = []

# loop through pages 0-19
for page in range(0,20):
    
    # create query with a page number
    # API results show 10 articles at a time
    page_query_url = f'{movie_url}&page={str(page)}'
    
    # Make a "GET" request and retrieve the JSON
    ten_movies = requests.get(page_query_url).json()
    
    # Add a twelve second interval between queries to stay within API query limits
    time.sleep(12)
    
    # Try and save the reviews to the reviews_list
    try:       
        # loop through the reviews["response"]["docs"] and append each review to the list
        for movie in range(0, 10):
            films = ten_movies['response']['docs'][movie]
            reviews_list.append(films)

        # Print the page that was just retrieved
            print(f'Page {page} Checked - Movie {movie} Checked')

        # Print the page number that had no results then break from the loop
    except KeyError:
        print(f'Page {page} had no results for movie {movie}')
        break
```

```python
# Preview the first 5 results in JSON format
# Use json.dumps with argument indent=4 to format data

print(json.dumps(reviews_list[0:5],indent = 4))
```

```python
# Convert reviews_list to a Pandas DataFrame using json_normalize()

reviews_list_df = pd.json_normalize(reviews_list)
reviews_list_df
```

```python
# Extract the title from the "headline.main" column and
# save it to a new column "title"
# Title is between unicode characters \u2018 and \u2019. 
# End string should include " Review" to avoid cutting title early

#NOTE TO THE GRADER: I tried a much more simplified approach before but the results were not great, the code below generetes a better list.

titles = []
end = 0

for x in range (0,200):
    text = reviews_list_df['headline.main'][x]
    y = len(text)    
    b = 0
    e = 0   
    
    if b <= y:
        word =''
        letter = ''       
        for begining_letter in text:         
            letter = text[b]
            word = word + letter
            b = b + 1
            if letter == '\u2018':
                break
        word =''
        letter = ''
        for ending_letter in text:
            letter = text[-e]
            word = word + letter
            e = e + 1
            if letter == '\u2019':
                break
    end = y-e+1
    title = text[b:end]
    titles.append(title)
reviews_list_df['title'] = titles
reviews_list_df
```

```python
# Extract 'name' and 'value' from items in "keywords" column

def extract_keywords(keyword_list):
    extracted_keywords = ""
    for item in keyword_list:
        # Extract 'name' and 'value'
        keyword = f"{item['name']}: {item['value']};" 
        # Append the keyword item to the extracted_keywords list
        extracted_keywords += keyword
    return extracted_keywords
    
reviews_list_df['keywords'] = reviews_list_df['keywords'].apply(extract_keywords)
```

```python
# Fix the "keywords" column by converting cells from a list to a string

pd.options.mode.copy_on_write = True
for values in range(0, 200):
    reviews_list_df['keywords'][values] = str(reviews_list_df['keywords'][values])
```

### Access The Movie Database API

```python
# Prepare The Movie Database query
url = "https://api.themoviedb.org/3/search/movie?query="
tmdb_key_string = "&api_key=" + tmdb_api_key
```

```python
# Create an empty list to store the results

tmdb_movies_list = []
relevant_data = {}
genre_list = []
languages = []
production_countries = []
counter = 0

# Create a request counter to sleep the requests after a multiple
# of 50 requests

# Loop through the titles

for movie in range(0,200):
    request_url = url + reviews_list_df['title'][movie] + tmdb_key_string   
    response_data = requests.get(request_url).json()
    
    # Check if we need to sleep before making a request
    if counter == 49: 
        time.sleep(60)
        counter = 0
    else:
        # Add 1 to the request counter
        counter = counter + 1
        # Perform a "GET" request for The Movie Database
        # Include a try clause to search for the full movie details.
        try:
            response_data = requests.get(request_url).json()
            
    # Use the except clause to print out a statement if a movie
    # is not found.
            # Get movie id
            movie_id = response_data['results'][0]['id']
        except IndexError:
            print(f"{movie} {Fore.YELLOW}NOT FOUND{Style.RESET_ALL}")
 
        # Make a request for a the full movie details           
        # Execute "GET" request with url 
        query_url = f'https://api.themoviedb.org/3/movie/{movie_id}?{tmdb_key_string}'
        full_movie_details = requests.get(query_url)
        details = full_movie_details.json()

        # Extract the genre names into a list
        y = len(details['genres'])
        for x in range(y):
            genre_list.append(details['genres'][x]['name'])       
        
        # Extract the spoken_languages' English name into a list
        y = len(details['spoken_languages'])
        for x in range(y):
            languages.append(details['spoken_languages'][x]['english_name'])
         
        # Extract the production_countries' name into a list
        y = len(details['production_countries'])
        for x in range(y):
            production_countries.append(details['production_countries'][x]['name'])

        # Add the relevant data to a dictionary and
        # append it to the tmdb_movies_list list  
        relevant_data['title'] = details['title'].replace("'","’")
        relevant_data['original_title'] = details['original_title']
        relevant_data['budget'] = details ['budget']
        relevant_data['genre'] = genre_list
        relevant_data['language'] = details ['original_language']
        relevant_data['spoken_languages'] = languages
        relevant_data['homepage'] = details ['homepage']
        relevant_data['overview'] = details ['overview']
        relevant_data['popularity'] = details ['popularity']
        relevant_data['runtime'] = details ['runtime']
        relevant_data['revenue'] = details ['revenue']
        relevant_data['release_date'] = details ['release_date']
        relevant_data['vote_average'] = details ['vote_average']
        relevant_data['vote_count'] = details ['vote_count']
        relevant_data['production_countries'] = production_countries
        
        tmdb_movies_list.append(relevant_data)
        
        # Print out the title that was found
        if reviews_list_df['title'][movie].lower() == relevant_data['title'].lower():
            print(f"{movie} {Fore.GREEN}FOUND{Style.RESET_ALL} {reviews_list_df['title'][movie]}")
            print(reviews_list_df['title'][movie].lower())
            print(details ['title'].lower()) 
        elif str(details['title'].lower()) == str(reviews_list_df['title'][movie].lower()):
            verdad = True
            print(verdad)
        else:
            print(f"{movie}{Fore.RED} NOT FOUND{Style.RESET_ALL} {reviews_list_df['title'][movie]}")
            print(reviews_list_df['title'][movie].lower())
            print(details ['title'].lower())
            
        genre_list = []
        languages = []
        production_countries = []
        relevant_data = {}
```

### Merge and Clean the Data for Export

```python
# Merge the New York Times reviews and TMDB DataFrames on title
merged_df = love_in_film_df.merge(reviews_list_df)
merged_df_clean = merged_df
```

```python
merged_df
```

```python
# Remove list brackets and quotation marks on the columns containing lists
# Create a list of the columns that need fixing
column_fix_list = []
column_fix_list = ['genre', 'spoken_languages']

# Create a list of characters to remove
remove_characters_list =[]
remove_characters_list = ["'", "[", "]"]
test_list_1 = []
fixed_list =[]

# # Loop through the list of columns to fix

for column in column_fix_list:
    fixed_list = merged_df[column]
    merged_df_clean[column] = fixed_list
    # Convert the column to type 'str'
    for item in fixed_list:
        item_string = str(item)
    # Loop through characters to remove
        for char in remove_characters_list:
            clean_item = item_string.replace(char,'')
            item_string = clean_item
        test_list_1.append(item_string)
    merged_df_clean[column] = test_list_1 
    test_list_1 = []

# Display the fixed DataFrame

merged_df_clean.head()
```

```python
# Drop "byline.person" column
merged_df_clean.drop('byline.person', axis = 1)
```

```python
# Delete duplicate rows and reset index
merged_df_final = merged_df_clean.drop_duplicates(subset=["title"], keep=False).reset_index(drop = True)
merged_df_final
```

```python
# Export data to CSV without the index
merged_df_final.to_csv('merged_df_final.csv', index = False)
```
